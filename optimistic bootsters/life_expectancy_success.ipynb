```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv("Life Expectancy Data.csv")
```

```python
data.isnull().sum().sort_values()
```

```python
def fill_with_mean(column, dataset=data):
  dataset[column].fillna(dataset[column].mean(), inplace=True)
```

```python
feature_missing = data.columns[data.isna().any()]

for column in feature_missing:
  fill_with_mean(column)
```

```python
#Columns with positive correlation from subset

positive_corr_pairs = sorted_pairs[sorted_pairs > 0]
print(positive_corr_pairs)
```
